## (2)设置属性，cmap调色板，linewidths网格的边界宽度

```python
x=sns.clustermap(flights,cmap='coolwarm',linewidths=1.4)
```

```python
x=sns.clustermap(flights,cmap='Blues_r',linewidths=1.4)
```

## （3）标准化或数值在score化，standard_scale,z_score,可True或0（列），1（行）

```python
x=sns.clustermap(flights,cmap='Blues_r',linewidths=1.4,standard_scale=0)
```

```python
x=sns.clustermap(flights,cmap='Blues_r',linewidths=1.4,standard_scale=1)
```

```python
x=sns.clustermap(flights,cmap='Blues_r',linewidths=1.4,z_score=True)
```

```python
x=sns.clustermap(flights,cmap='Blues_r',linewidths=1.4,z_score=0)
```

```python
x=sns.clustermap(flights,cmap='Blues_r',linewidths=1.4,z_score=1)
```
