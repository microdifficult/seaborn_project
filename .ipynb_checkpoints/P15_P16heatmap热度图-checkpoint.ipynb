```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data=np.random.randn(12,15)
```

```python
plt.figure(figsize=(8,8))
x=sns.heatmap(data,annot=True,vmin=0,vmax=2)
```

```python
flights=sns.load_dataset('flights')
flights=flights.pivot('month','year','passengers')
```

```python
plt.figure(figsize=(9,9))
x=sns.heatmap(flights,annot=True,fmt='d',cmap='summer',linewidths=1.1)
```

```python
plt.figure(figsize=(9,9))
sns.heatmap(flights,annot=True,fmt='d',linewidths=1.1,center=flights.loc['July',1954])
```
