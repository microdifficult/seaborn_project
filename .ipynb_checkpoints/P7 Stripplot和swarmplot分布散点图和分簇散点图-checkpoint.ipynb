# stripplot和swarmplot是对violinplot和boxplot的补充，常用于相互叠合

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips=sns.load_dataset("tips")
```

# （1）比较分布散点图和分簇散点图的视觉效果的差异，用法和barplot，boxplot相同

```python
sns.stripplot(x='total_bill',data=tips)
```

```python
sns.swarmplot(x='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

# （2）参数jitter使点更离散，可设置成bool或设置抖动值

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=0.08)
```

# （3）参数linewidth设置点大小

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=0.08,linewidth=0.5)
```

```python
sns.stripplot(x='day',y='total_bill',hue='sex',data=tips,jitter=0.08,linewidth=0.5)
```

# （4）参数dodge设置hue分离

```python
sns.stripplot(x='day',y='total_bill',hue='sex',data=tips,jitter=0.08,linewidth=0.5,dodge=True)
```

# （5）参数marker,size,edgecolor设置样点的样式

```python
sns.stripplot(x='day',y='total_bill',hue='sex',data=tips,jitter=0.08,linewidth=0.5,dodge=True,marker='D',size=3,edgecolor='red')
```

# （6）参数alpha设置透明度只对点数据有效

```python
sns.violinplot(x='day',y='total_bill',data=tips,color='0.8')
sns.stripplot(x='day',y='total_bill',data=tips,jitter=0.2,alpha=0.2)
```
