```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris=sns.load_dataset("iris")
```

# （1）map函数详解：x.map(函数)，x数据集，函数方法

# （2）常以赋值形式画图，把图作为变量成为画图句柄

```python
x=sns.PairGrid(iris)
x=x.map(plt.scatter)
```

```python
iris.head()
```

# （3）控制pairplot位置画图的几大map：map_diag（对线）,map_offdiag（对

# 角线外）,map_lower（下三角）,map_upper（上三角）

```python
x=sns.PairGrid(iris,hue='species')
x=x.map_diag(plt.hist)
x=x.map_offdiag(sns.kdeplot)
```

```python
help(sns.PairGrid)
```

# （4）hue是在生成数据里即在PairPlot中形成的，并非map的参数

# （5）x=x.add_legend()，x设为图像的好处可以随时进行设置

```python
x=sns.PairGrid(iris,hue='species')
x=x.map_diag(plt.hist)

x=x.map_upper(plt.scatter)
x=x.map_lower(sns.kdeplot)
x=x.add_legend()
```

```python
help(sns.kdeplot)
```

# （6）PairGrid可控制变量，设置处理的变量，vars=[]

```python
x=sns.PairGrid(iris,hue='species',vars=['sepal_length','sepal_width'])
x=x.map(plt.scatter)
```

# （7）PairGrid可控制x,y变量，设置处理的变量，x_vars=[],y_vars=[]

```python
x=sns.PairGrid(iris,hue='species',x_vars=['sepal_length','sepal_width'],y_vars=['petal_length','petal_width'])
x=x.map(plt.scatter)
x=x.add_legend()
```

# ￥￥￥￥_kws:keywords关键词多用于样式设计

```python
x=sns.PairGrid(iris,hue='species',vars=['sepal_length','sepal_width'],hue_kws={'marker':['D','s','+']})
x=x.map(plt.scatter)
```
