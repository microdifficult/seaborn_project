```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

# (1)用法和boxplot一模一样

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

## （2）hue和split选项契合度高，split上个有此选项的为stripplot和swarmplot（dodge）

```python
x=sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

# （3）inner图内有四分位线，scale默认为频率，改为频数

```python
x=sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True,inner='quartile',scale='count')
```

```python
x=sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True,inner='stick',scale='count')
```

# （4）scale_hue:小提琴不标准化

```python
x=sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True,inner='stick',scale='count',scale_hue=False)
```

# （5）bw:拟合度

```python
x=sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True,inner='stick',scale='count',scale_hue=False,bw=0.2)
```
