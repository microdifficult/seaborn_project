```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

# （1）两条语句必须挨着不能分开，否则报错

# （2）要用赋值语句画图

```python
x= sns.FacetGrid(tips,row='sex',col='smoker',hue='time')
x=x.map(plt.scatter,'total_bill','tip')
```

```python
x= sns.FacetGrid(tips,row='sex',col='smoker')
x=x.map(plt.hist,'total_bill',bins=15,color='Red').add_legend()
```

```python
x= sns.FacetGrid(tips,row='sex',col='smoker',hue='time')
x=x.map(sns.regplot,'total_bill','tip').add_legend()
```

# （3）格式问题，画布格式在grid中，aspect横纵比，交会图格式，在map内

```python
x= sns.FacetGrid(tips,row='sex',col='smoker',hue='time',size=5,aspect=1)
x=x.map(sns.regplot,'total_bill','tip').add_legend()
```
