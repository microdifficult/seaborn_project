# 条形图是分类变量的统计信息，即分类变量vs连续变量colorwith分类变量

# 利用help（函数）进行函数参数知晓

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns


%matplotlib inline
```

# （1）sns.load_dataset加载在线数据集

```python
tips=sns.load_dataset("tips")
```

```python
tips.info()
```

```python
tips.head(100)
```

# （2）sns.barplot条形图，输入xy，data是数据集

```python
sns.barplot(x="day",y="tip",data=tips)
```

```python
sns.barplot(x='day',y='total_bill',data=tips)
```

# （3）sns.barplot条形图，hue是加入第三分类变量显示

```python
sns.barplot(x='day',y='total_bill',data=tips,hue='sex')
```

# （4）sns.barplot条形图，palette调色板，color纯色板

```python
sns.barplot(x='day',y='total_bill',data=tips,hue='sex',palette='winter_r')
```

```python
sns.barplot(x='day',y='total_bill',data=tips,hue='sex',color='red')
```

# （5）sns.barplot条形图，x，y调节方向

```python
sns.barplot(x='total_bill',y='day',data=tips)
```

# （6）sns.barplot条形图，order=['','']调节x轴分类变量顺序

```python
sns.barplot(x='day',y='total_bill',data=tips,order=['Fri','Thur','Sat','Sun'])
```

# （7）sns.barplot条形图，estamator调整y的计算方式，平均值改为中值，中值

# 利用需要重新调入numpy

```python
from numpy import median
sns.barplot(x='day',y='total_bill',data=tips,estimator=median)
```

```python
help(sns.barplot)
```

# （8）sns.barplot条形图，ci=  误差条,capsize=误差条的帽子

```python
sns.barplot(x='smoker',y='total_bill',data=tips,ci=99)
```

```python
sns.barplot(x='smoker',y='total_bill',data=tips,ci=100,capsize=0.3)
```
