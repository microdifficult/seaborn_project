```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
iris=sns.load_dataset('iris')
```

```python
tips.head()
```

```python
iris.head()
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips)
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris)
```

 #  kind must be either 'scatter', 'reg', 'resid', 'kde', or 'hex'


```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris,kind='kde')
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg',color='red')
```

```python
help(sns.jointplot)
```

```python
from scipy.stats import spearmanr
```

# stat_func:spearmanr相关系数

```python
sns.jointplot(x='size',y='total_bill',data=tips,stat_func=spearmanr,color='red')
```

```python
sns.jointplot(x='size',y='total_bill',data=tips,color='red')
```

# 样式设计size,ratio

```python
sns.jointplot(x='size',y='total_bill',data=tips,stat_func=spearmanr,color='red',size=5,ratio=5)
```
