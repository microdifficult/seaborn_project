```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips =sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.regplot(x='tip',y='total_bill',data=tips,fit_reg=True)
```

```python
mean=[2,5]
cov=[[1.1,0.4],[2.2,3]]
x_value,y_value=np.random.multivariate_normal(mean,cov,100).T
```

```python
sns.regplot(x=x_value,y=y_value)
```

```python
Series_x_value=pd.Series(x_value,name='Var_x')
Series_y_value=pd.Series(y_value,name='Var_y')
```

# 属性：color，marker

```python
sns.regplot(x=Series_x_value,y=Series_y_value,color='red', marker='D')
```

## 点属性，线属性scatter_kws={'color':"R",},line_kws={'color':"blue","linewidth":1.5}

```python
sns.regplot(x=Series_x_value,y=Series_y_value,ci=68,marker='D',scatter_kws={'color':"R",},line_kws={'color':"blue","linewidth":1.5})
```

```python
sns.regplot(x='size',y='total_bill',data=tips)
```

# 抖动x_jitter=0.1，适用于离散变量，x_estimator=np.median，设置预测值

```python
sns.regplot(x='size',y='total_bill',data=tips,x_jitter=0.1)
```

```python
sns.regplot(x='size',y='total_bill',data=tips,x_jitter=0.1,x_estimator=np.median)
```
