```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.pairplot(iris,hue='species')
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

# 属性hue, markers（对应不同hue标记），size


```python
sns.pairplot(tips,hue='sex',markers=['+','D'])
```

# vars（变量）,x_vars,y_vars

```python
sns.pairplot(tips,hue='sex',markers=['+','D'],vars=['size','total_bill'],height=4)
```

```python
sns.pairplot(iris,hue='species',markers=['+','D','o'],x_vars=['sepal_length','sepal_width'],\
             y_vars=['petal_length','petal_width'],height=4)
```

# diag_kind:对角线画图格式，kind回归线

```python
sns.pairplot(iris,hue='species',diag_kind='kde',height=4,kind='reg')
```
