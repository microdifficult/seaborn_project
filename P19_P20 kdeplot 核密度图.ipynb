# kdeplot具有部分交会图的功能

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
mean=[0,0]
cov=[[0.2,0],[0,3]]
```

```python
x_axis,y_axis=np.random.multivariate_normal(mean,cov,size=40).T
```

```python
sns.kdeplot(x_axis)
```

```python
sns.kdeplot(y_axis)
```

```python
sns.kdeplot(x_axis,shade=True)
```

```python
sns.kdeplot(x_axis,y_axis,cmap='RdBu')
```

# （1）kdeplot:shade 阴影，n_level等值线个数，cmp图版

```python
sns.kdeplot(x_axis,y_axis,shade=True,n_levels=18,cmap='coolwarm')
```

```python
sns.kdeplot(x_axis,shade=True,bw=0.5)
```

```python
sns.kdeplot(x_axis,shade=True,bw=0.01)
```

```python
sns.kdeplot(x_axis,shade=True,bw=0.1)
```

```python
iris=sns.load_dataset('iris')
```

```python
setosa=iris.loc[iris.species=='setosa']
versicolor=iris.loc[iris.species=='versicolor']
```

```python
sns.kdeplot(setosa.petal_length,setosa.petal_width)
sns.kdeplot(versicolor.petal_length,versicolor.petal_width)
```
